```python
# Write code in code blocks.

import numpy as np
import scipy.stats
from scipy.stats import norm
import matplotlib as mp
import matplotlib.pyplot as plt
```

Generate m = 1000 independent draws of different sample spaces.

```python
n=20
sqrtn=np.sqrt(n)
m=1000
xnorm=[]
xpoi=[]
xunif=[]
xber=[]
np.random.seed(1)
for i in range(m):
    xnorm.append(sqrtn*np.mean(np.random.normal(0,1,n)))
    xber.append(sqrtn*np.mean(np.random.binomial(1,0.5,n)*2-1))
    xpoi.append(sqrtn*np.mean(np.random.poisson(1,n)-1))
    xunif.append(sqrtn*np.mean(np.sqrt(12)*(np.random.uniform(0,1,n)-0.5)))
```

(a)

Write a function generate random walk.

```python
def generate_random_walk(p,n):
    w = [0]
    steps = 2 * np.random.binomial(1,p,n) - 1
    wsum = 0
    for i in range(n):
        wsum = wsum + steps[i]
        w.append(wsum)
    return np.array(w)
```

(b)

Make a plot that shows the trajectories of five different independent random walks generated from the function.

```python
n = 200
plt.title('random walk')
plt.xlabel('$i$')
plt.ylabel('$W_i$')
plt.plot(generate_random_walk(0.4,n),'r',label='$p=0.4$')
plt.plot(generate_random_walk(0.45,n),'g',label='$p=0.45$')
plt.plot(generate_random_walk(0.5,n),'b',label='$p=0.5$')
plt.plot(generate_random_walk(0.55,n),'c',label='$p=0.55$')
plt.plot(generate_random_walk(0.6,n),'m',label='$p=0.6$')
_ = plt.legend(loc = 'best')
```
